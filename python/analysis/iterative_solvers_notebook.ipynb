# Plotting

## Processing Data for Plots

```python
# ['jacobi', 'Gauß-Seidel', 'gradient', 'conjugate_gradient']
## ['iterations', 'matrix_dimension', 'tol', 'max_iter', 'err_rel', 'iterations_number', 'elapsed_time']
PARAMS = ['err_rel', 'iterations_number', 'elapsed_time', 'iteration_time_avg']

def processing_parameters_data(exec_data):
    data_plot = {}
    data_plot['param'] = []
    for method in exec_data.keys():
        data_plot[method] = []
        for param in PARAMS:
            if len(data_plot['param']) < len(PARAMS):
                data_plot['param'].append(param)
            data_plot[method].append(exec_data[method][param])


    df_plot = pd.DataFrame(data_plot)
    df_plot.head()
    
    return df_plot
```

## Plotting

```python
def plotting_params(df_plot, matrix, tol):
    for param in df_plot['param'].unique():

        df_plot[df_plot['param']== param].plot(kind='bar')

        plt.title(f"param = {param}\nmatrix = {matrix}\ntolerance = {tol}")

        plt.plot()
```

## Plotting Convergence

```python
# ['jacobi', 'Gauß-Seidel', 'gradient', 'conjugate_gradient']
def plotting_convergence(exec_data, matrix, tol):
    for method in exec_data.keys():
        plt.figure(figsize=(10, 5))
        
        plt.subplot(1, 2, 1)
        plt.title(f"{method} convergence:\nmatrix = {matrix} \ntolerance = {tol}")
        plt.plot(exec_data[method]['iterations'].keys(), exec_data[method]['iterations'].values(), color='blue')
        
        
        plt.subplot(1, 2, 2)
        abs_values = [abs(val) for val in exec_data[method]['iterations'].values()]
        plt.title(f"{method} absolute convergence:\nmatrix = {matrix} \ntolerance = {tol}")
        plt.plot(exec_data[method]['iterations'].keys(), abs_values, color='blue')
```

### Tempo medio a iterazione?

```python
for matrix in exec_data.keys():
    for tol in exec_data[matrix].keys():
        df_plot = processing_parameters_data(exec_data[matrix][tol])
        plotting_params(df_plot, matrix, tol)
```

```python
for matrix in exec_data.keys():
    for tol in exec_data[matrix].keys():
        plotting_convergence(exec_data[matrix][tol], matrix, tol)
```
